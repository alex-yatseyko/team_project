```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data for Keras model
keras_confusion_matrix = np.array([[6365, 4], [1731, 46]])
keras_precision = [0.79, 0.92]
keras_recall = [1.00, 0.03]
keras_f1_score = [0.88, 0.05]
keras_accuracy = 0.79

# Data for PyTorch model
torch_confusion_matrix = np.array([[4851, 221], [474, 971]])
torch_precision = [0.91, 0.81]
torch_recall = [0.96, 0.67]
torch_f1_score = [0.93, 0.74]
torch_accuracy = 0.89
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6)) # Plotting confusion matrices

sns.heatmap(keras_confusion_matrix, annot=True, fmt='d', cmap='Blues', ax=ax[0])
ax[0].set_title('Keras Model Confusion Matrix')
ax[0].set_xlabel('Predicted Label')
ax[0].set_ylabel('True Label')

sns.heatmap(torch_confusion_matrix, annot=True, fmt='d', cmap='Greens', ax=ax[1])
ax[1].set_title('PyTorch Model Confusion Matrix')
ax[1].set_xlabel('Predicted Label')
ax[1].set_ylabel('True Label')

plt.show()
```

```python
# Bar charts for Precision, Recall, and F1-Score
labels = ['Class 0', 'Class 1']
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Precision
ax[0].bar(x - width/2, keras_precision, width, label='Keras', color='blue')
ax[0].bar(x + width/2, torch_precision, width, label='PyTorch', color='green')
ax[0].set_title('Precision Comparison')
ax[0].set_xticks(x)
ax[0].set_xticklabels(labels)
ax[0].legend()

# Recall
ax[1].bar(x - width/2, keras_recall, width, label='Keras', color='blue')
ax[1].bar(x + width/2, torch_recall, width, label='PyTorch', color='green')
ax[1].set_title('Recall Comparison')
ax[1].set_xticks(x)
ax[1].set_xticklabels(labels)
ax[1].legend()

# F1-Score
ax[2].bar(x - width/2, keras_f1_score, width, label='Keras', color='blue')
ax[2].bar(x + width/2, torch_f1_score, width, label='PyTorch', color='green')
ax[2].set_title('F1-Score Comparison')
ax[2].set_xticks(x)
ax[2].set_xticklabels(labels)
ax[2].legend()

plt.show()
```

```python
# Accuracy comparison
fig, ax = plt.subplots(figsize=(6, 6))

ax.bar(['Keras', 'PyTorch'], [keras_accuracy, torch_accuracy], color=['blue', 'green'])
ax.set_title('Accuracy Comparison')
ax.set_ylabel('Accuracy')

plt.show()
```
